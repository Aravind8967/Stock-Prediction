```python
from src import CompanyDetails, SharePricePrediction
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
c_name = 'TATAMOTORS'
data = SharePricePrediction(c_name)

prediction = data.prediction(pred_range='6y')

print(prediction)
```
